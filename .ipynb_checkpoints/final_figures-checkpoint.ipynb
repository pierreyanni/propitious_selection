# Using pr_nhome_compas
We adjust the probability of nursing home to be smaller or equal to the probablity of probability of survival to 85 years old.

```python
df = pd.read_excel("..\data\compas_enquete_mll.xlsx")
df.rename(columns={'pr_nhome_compas': 'alpha1',
                   'pr_alive_85_compas': 'pi'},
          inplace=True)

print(f'prob dependency before 85: {(df.alpha1 > df.pi).mean()}')
df['alpha1'] = df[['alpha1', 'pi']].min(axis='columns')
df['alpha2'] = df.pi - df.alpha1
df['death'] = 1 - df.alpha1 - df.alpha2
df = df[['alpha1', 'alpha2', 'death']]
df.describe()
```

## Find propitious selection

```python
w = 100e3

l_propitious_nhome = []
for L1 in np.arange(60e3, w + 1, 1000):
    for L2 in np.arange(40e3, L1, 1000):
        propitious = Propitious(w, df.alpha1, df.alpha2, L1, L2)
        if propitious.check_propitious_selection():
            l_propitious_nhome.append((L1, L2))
plot_propitious_selection(l_propitious_nhome)
```

## Find prices (and profits) for which there is propitious selection
For each pair (L1, L2) for which we found propitious selection, the profit maximizing price and the prices for which there is propitious selection (average cost over the next price increase) are shown, as well as associated profits. Prices are sorted in increasing order.

```python
for L1, L2 in l_propitious_nhome:
    propitious = Propitious(w, df.alpha1, df.alpha2, L1, L2)
    propitious_prices = propitious.find_prices_profits_propitious_selection(step=50)
    print(f'L1={L1}, L2={L2}')
    display(propitious_prices)    
```

## Specific example of propitious selection
We pick three pairs (L1, L2) from the list above and show all the prices for which propitious selection occurs.

```python
for L1, L2 in [(86000, 55000), (87000, 55000), (93000, 62000)]:
    propitious = Propitious(w, df.alpha1, df.alpha2, L1, L2)
    _, prices = propitious.find_prices_propitious_selection(step=50)
    for price in prices:
        print(f'L1={L1}, L2={L2}')
        print(f'price={price}')
        propitious.create_figure(price)
```

## Cosmetics
Some potential improvements to the figures

### Basic version

```python
L1, L2 = (93000, 62000)
P = 66350
print(f'L1={L1}, L2={L2}, price={P}')
propitious = Propitious(w, df.alpha1, df.alpha2, L1, L2)
propitious.calc_UP_A1_A2(P)
UP, A1, A2 = propitious.UP, propitious.A1, propitious.A2
C = propitious.average_cost(P)

alpha1pref = [alpha1 for alpha1 in np.linspace(0, UP/A1, 100)]
alpha2pref = [UP/A2 - alpha1 * A1/A2 for alpha1 in alpha1pref]

C = propitious.average_cost(P)
alpha1cost = [alpha1 for alpha1 in np.linspace(0, C/L1, 100)]
alpha2cost = [C/L2 - alpha1 * L1/L2 for alpha1 in alpha1cost]

plt.scatter(df.alpha1, df.alpha2, alpha=0.1)
plt.plot(alpha1pref, alpha2pref, label='isopreference')
plt.plot(alpha1cost, alpha2cost, label='isocost')
plt.xlabel(r'$\alpha_1$')
plt.ylabel(r'$\alpha_2$')
plt.grid()
plt.legend()
plt.show()
```

### Participation/non-participation

```python
L1, L2 = (93000, 62000)
P = 66350
print(f'L1={L1}, L2={L2}, price={P}')
propitious = Propitious(w, df.alpha1, df.alpha2, L1, L2)
propitious.calc_UP_A1_A2(P)
UP, A1, A2 = propitious.UP, propitious.A1, propitious.A2
C = propitious.average_cost(P)

alpha1pref = [alpha1 for alpha1 in np.linspace(0, UP/A1, 100)]
alpha2pref = [UP/A2 - alpha1 * A1/A2 for alpha1 in alpha1pref]

C = propitious.average_cost(P)
alpha1cost = [alpha1 for alpha1 in np.linspace(0, C/L1, 100)]
alpha2cost = [C/L2 - alpha1 * L1/L2 for alpha1 in alpha1cost]

participant = propitious.participation(P)

plt.scatter(df.alpha1[participant], df.alpha2[participant], alpha=0.1,
            color='forestgreen')
plt.scatter(df.alpha1[~participant], df.alpha2[~participant], alpha=0.2,
            color='indianred')
plt.plot(alpha1pref, alpha2pref, label='isopreference', color='blue')
plt.plot(alpha1cost, alpha2cost, label='isocost', color='black')
plt.xlabel(r'$\alpha_1$')
plt.ylabel(r'$\alpha_2$')
plt.grid()
plt.legend()
plt.show()
```
